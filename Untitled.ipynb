```python
# Imports
from datetime import datetime, date

import pandas as pd
from pandas.tseries.offsets import BDay
```

```python
pd.set_option('expand_frame_repr',False)
pd.set_option('display.max_rows',1000)
```

```python
import requests
url1='https://api.huobi.br.com/market/history/kline?period=60min&size=2000&symbol=btcusdt'
url2='https://api.huobi.br.com/market/history/kline?period=60min&size=2000&symbol=ethusdt'
url3='https://api.huobi.br.com/market/history/kline?period=60min&size=2000&symbol=ltcusdt'
resp1=requests.get(url1)
resp2=requests.get(url2)
resp3=requests.get(url3)
resp_json1=resp1.json()
resp_json2=resp2.json()
resp_json3=resp3.json()
data_list1=resp_json1['data']
data_list2=resp_json2['data']
data_list3=resp_json3['data']
df1=pd.DataFrame(data_list1)
df2=pd.DataFrame(data_list2)
df3=pd.DataFrame(data_list3)
dff=df1.merge(df2, how='left', on='id').merge(df3, how='left', on='id')
dff
```

```python
dff.drop(['open_x', 'low_x','amount_x','high_x','count_x','vol_x',
         'open_y', 'low_y','amount_y','high_y','count_y','vol_y',
         'open', 'low','amount','high','count','vol',], axis=1, inplace=True)
dff
```

```python
# getting relevant dates
```

```python
#rename
dff.rename({'close': 'LTC', 'close_x': 'BTC','close_y': 'ETH'}, axis=1, inplace=True)
dff['date']=pd.to_datetime(dff['id'], unit='s')
res = dff[~(dff['date'] > '2022-12-23')]
res=res.set_index(['date'])
res.drop(['id',], axis=1, inplace=True)
res
```

Return estimation

```python
from pypfopt import expected_returns
mu = expected_returns.capm_return(res)
mu.plot.barh(figsize=(10,6));
```

Long/short min variance

```python
from pypfopt import EfficientFrontier

S = risk_models.CovarianceShrinkage(res).ledoit_wolf()

# You don't have to provide expected returns in this case
ef = EfficientFrontier(None, S, weight_bounds=(None, None))
ef.min_volatility()
weights = ef.clean_weights()
weights
```

```python
pd.Series(weights).plot.barh();
```

plotting Efficient frontier

```python

from pypfopt import CLA, plotting

cla = CLA(mu, S)
cla.max_sharpe()
cla.portfolio_performance(verbose=True);
```

```python
ax = plotting.plot_efficient_frontier(cla, showfig=True)
```
